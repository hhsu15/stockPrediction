## PCA

from Wikipedia: *Principal component analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert a set of observations of possibly correlated variables into a set of values of linearly uncorrelated variables called principal components.*

In English: Principal component analysis transforms the original variables in to new variables where the new variables are linear combinations of the original variables. These new variables will be uncorrelated (linearly). The new variables (Principal Components) are ordered by how much variance in the data they explain. This allows us to keep the top few Principal Components while still keeping most of the information contained in the data.

You can also think of PCA as a axes rotation.

PCA was developed in 1901!

### PCA with the Iris Dataset
http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# load and plot dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data,
                       columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]
sns.pairplot(iris_df, hue='species');
```

```python
# perform PCA and pair plot
X = iris_df.drop('species', axis=1)
pca = PCA()
X_pca = pca.fit_transform(X)

iris_pca_df = pd.DataFrame(data=X_pca,
                           columns=['PCA_' + str(i) for i in range(pca.n_components_)])
iris_pca_df['species'] = iris.target_names[iris.target]

sns.pairplot(iris_pca_df, hue='species');
```

```python
iris_pca_df
```

```python
# variance explained by Principal Components
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel('number of PCs')
plt.ylabel('variance explained')
```

### Next steps
The new features (Principal Components) can then be used to perform classification/regression. For this example, it's unlikely to make a difference in accuracy, but for a problem that has a lot more features (100s, 1000s or more!), this technique can dramatically improve the results.
